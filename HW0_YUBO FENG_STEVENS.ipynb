```python
from urllib.request import urlopen
import requests

global_headers = {'user-agent': 'a health spdder'}
```

```python
def create_search_url(data):
    from urllib.parse import urlencode
    key_word_pair = urlencode(data)
    return 'https://www.google.com/' + 'search?' + key_word_pair

data = {'q':'Tim Berners-Lee'}
url1 = create_search_url(data)

print(url1)
```

# Question 1
* The body of google search result
* The status of google search result
* the returned headers

```python
sess = requests.Session()
response = sess.get(url1, headers = global_headers)
print(response.text, '\n')
print(response.status_code, '\n')
print(response.headers, '\n')
```

# Question 2
* The body of post
* status
* headers

```python
url = 'http://www.cplusplus.com/'
data = {'ver':'2.1'}
response = requests.post(url, headers=global_headers, data=data)
print(response.text, '\n')
print(response.status_code, '\n')
print(response.headers, '\n')
```
